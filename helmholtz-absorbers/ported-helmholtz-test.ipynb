# Nearfield to the device ~45d angle

<img src="../images/helmhotlz-ported-device-mic.jpg" width="400">

```python
def compare_measurements(a, b, title, vline=None):
    a_data = measurements[a]
    freq = a_data["freq"]
    skip = len(freq[freq <= 30])
    ax1, ax2 = compare(
        a_data["freq"][skip:],
        a,
        a_data["spl"][skip:],
        b,
        measurements[b]["spl"][skip:],
        f"{title} - SPL",
    )
    if vline is not None:
        ax1.axvline(vline)
```

```python
compare_measurements("p2_device_84hz", "p2_device_closed", "84Hz", vline=84)
```

```python
compare_measurements("p2_device_102hz", "p2_device_closed", "102Hz", vline=102.0)
```

```python
compare_measurements(
    "p2_device_102hz-hard", "p2_device_closed", "102Hz Hard", vline=102.0
)
```

```python
compare_measurements(
    "p2_device_102hz-hard", "p2_device_102hz", "102Hz Hard vs Rounded", vline=102.0
)
```
