```python
import matplotlib.pyplot as plt
import numpy as np

from acoustics.constants import speed_of_sound
```

```python
t = np.arange(0.001, 0.021, 0.001)
d = speed_of_sound * t
```

```python
fig = plt.figure()
fig.set_figheight(6)
fig.set_figwidth(12)
ax = fig.add_subplot()
ax.set_title("Travel Time (meters)")

ax.plot(t, d);
```

```python
fig = plt.figure()
fig.set_figheight(6)
fig.set_figwidth(12)
ax = fig.add_subplot()
ax.set_title("Travel Time (feet)")

ax.plot(t, d / 0.304);
```
